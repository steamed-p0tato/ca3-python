```python
crossTAb = pd.crosstab(df['PLAYER NAME'], df['COUNTRY'])
print(crossTAb)
```

```python
sortedDf = df.sort_values(by='SOLD PRICE', ascending=False)
print(sortedDf)
```

```python
df['PREMIUM'] = df['SOLD PRICE'] - df['BASE PRICE']
maxPremiumPlayer = df.loc[df['PREMIUM'].idxmax()]
print(maxPremiumPlayer)
```

```python
max_premium = df['PREMIUM'].max()
players_max_premium = df[df['PREMIUM'] == max_premium]
print(players_max_premium)
```

```python
average_sold_price_age = df.groupby('AGE')['SOLD PRICE'].mean()
print(average_sold_price_age)
```

```python
average_sold_price_role_age = df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean()
print(average_sold_price_role_age)
```
